```python
import os
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline 
```

```python
file = open("./policyant2.out", "r")
```

```python
episode = []
steps = []
rewards = []

for line in file:
    v = line.strip(':').split()
    if len(v)==10:
        episode.append(int(v[3].strip(":")))
        rewards.append(float(v[6]))
        steps.append(int(v[8]))

print(len(episode))
print(max(episode))
print(len(steps))
print(max(steps))
print(len(rewards))
print(max(rewards))
```

```python
t = 10000
plt.plot(episode[:t], steps[:t])
plt.xlabel('Episode')
plt.ylabel('Steps')
plt.title('Steps per Episode')
plt.show()
```

```python

plt.plot(episode[:t], rewards[:t])
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Reward per Episode')
plt.show()
```
