```python
import os
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline 
```

```python
file = open("test10k.out", "r")
```

```python
episode = []
steps = []
rewards = []

for line in file:
    v = line.strip(':').split()
#     print(v)
    if len(v)==9:
        episode.append(int(v[3].strip(":")))
        rewards.append(float(v[5].split("=")[1]))
#         rewards.append(float(v[6]))
        steps.append(int(v[7]))

print(len(episode))
print(max(episode))
print(len(steps))
print(max(steps))
print(len(rewards))
print(max(rewards))
```

```python
t = 10000
plt.plot(episode[:t], steps[:t])
plt.xlabel('Episode')
plt.ylabel('Steps')
plt.title('Steps per Episode')
plt.show()
```

```python
rewards = np.array(rewards)*-1

plt.plot(episode[:t], rewards[:t])
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Actor-Critic with Learning Rates .001, .000001')
plt.show()
```
